   Добавляю реакцию:  "Glucose-6-phosphate + NADP+ + H2O to Gluconolactone + NADPH + H+"

```python
# Создание метаболитов
glucose6p = Metabolite("g6p_c", formula="C6H13O9P", name="Glucose-6-phosphate", compartment="c")
nadp_plus = Metabolite("nadp_c", formula="C21H25N7O17P3", name="NADP+", compartment="c")
h2o = Metabolite("h2o_c", formula="H2O", name="Water", compartment="c")
gluconolactone = Metabolite("glcn_c", formula="C6H9O6", name="Glucanolactone", compartment="c")
nadph = Metabolite("nadph_c", formula="C21H26N7O17P3", name="NADPH", compartment="c")
h_plus = Metabolite("h_c", formula="H", name="H+", compartment="c")
```

```python
reaction_id = "G6P_NADP_H2O_to_GLCN_NADPH_H"
reaction_name = "Glucose-6-phosphate + NADP+ + H2O to Gluconolactone + NADPH + H+"
reaction = Reaction(reaction_id)
reaction.name = reaction_name
```

```python
reaction.add_metabolites({glucose6p: -1, nadp_plus: -1, h2o: -1, gluconolactone: 1, nadph: 1, h_plus: 1})
```

```python
for metabolite in [glucose6p, nadp_plus, h2o, gluconolactone, nadph, h_plus]:
    if metabolite not in model.metabolites:
        model.add_metabolites([metabolite])
```

```python
model.add_reactions([reaction])
```

```python
print(f"Добавлена реакция: {reaction_id} - {reaction_name}")
print(f"Реакция добавлена в модель: {reaction_id in model.reactions}")
```
